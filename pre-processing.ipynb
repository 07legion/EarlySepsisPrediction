```python
import os
import pandas as pd
import csv
import numpy as np
pd.set_option('display.max_rows', 1500)
pd.set_option('display.max_columns', 1500)
```

```python
vitals=['HR','O2Sat','Temp','SBP','MAP','DBP','Resp','EtCO2']
demogs=['Age','Gender','Unit1','Unit2','HospAdmTime','ICULOS']
labs=['BaseExcess','HCO3','FiO2','pH','PaCO2','SaO2','AST','BUN','Alkalinephos','Calcium','Chloride','Creatinine','Bilirubin_direct','Glucose','Lactate','Magnesium','Phosphate','Potassium','Bilirubin_total','TroponinI','Hct','Hgb','PTT','WBC','Fibrinogen','Platelets']
```

```python
#compile all results
dirr='/Users/bilals01/Documents/time_series/project/training_setB'

folders = [f for f in sorted(os.listdir(dirr))]

vitals_df = pd.read_csv(dirr+'/'+folders[0], sep = "|")[vitals]
demogs_df = pd.read_csv(dirr+'/'+folders[0], sep = "|")[demogs]
labs_df = pd.read_csv(dirr+'/'+folders[0], sep = "|")[labs]
vitals_df['pat_id']=folders[0][1:7]
demogs_df['pat_id']=folders[0][1:7]
labs_df['pat_id']=folders[0][1:7]

for f in folders:
    print(f)
    df = pd.read_csv(dirr+'/'+f, sep = "|")
    v= df[vitals]
    d = df[demogs]
    l = df[labs]
    v['pat_id']=f[1:7]
    d['pat_id']=f[1:7]
    l['pat_id']=f[1:7]
    vitals_df=vitals_df.append(v)
    demogs_df=demogs_df.append(d)
    labs_df=labs_df.append(l)
    
```

```python
vitals_df=vitals_df.reset_index()
labs_df=labs_df.reset_index()
demogs_df=demogs_df.reset_index()

vitals_df=vitals_df.rename(columns={'index':'hour'},inplace=True)
labs_df=labs_df.rename(columns={'index':'hour'},inplace=True)
demogs_df=demogs_df.rename(columns={'index':'hour'},inplace=True)


labs_df
```
